```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
```

```python
img_path = 'C:\\Users\\duyson\\Desktop\\Projects\\FaceNormalize\\PytorchGAN\\dataset\\cfp-dataset\\Data\\Images\\001\\frontal\\02.jpg'
landmark_path = 'C:\\Users\\duyson\\Desktop\\Projects\\FaceNormalize\\PytorchGAN\\dataset\\cfp-dataset\\Data\\Fiducials\\001\\frontal\\02.txt'
img = misc.imread(img_path)
locations = np.loadtxt(landmark_path, delimiter=',')
locations = np.transpose(locations)
x = locations[0]
y = locations[1]
plt.figure()
plt.imshow(img)
plt.scatter(x, y, s = 50, c='red', marker='o')
plt.show()
```
